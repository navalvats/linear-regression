$(c)$ Determine where $f$ is increasing and where it is decreasing. Find the exact values where $f$ has a relative
maximum/minimum.

```python
from sympy import Poly
from sympy.abc import x
from sympy.solvers.inequalities import solve_poly_inequality
```

Increasing

```python
solve_poly_inequality(Poly(3*(x**2) - 8*x , x, domain='ZZ'), '>=')
```

Decreasing

```python
solve_poly_inequality(Poly(3*(x**2) - 8*x , x, domain='ZZ'), '<=')
```

Relative Maximum/Minimum

```python
from sympy import *
x = Symbol('x')
f = x**3 - 4*x**2 + 1
df = diff(f)
ddf = diff(df)
Roots = solveset(f, x)
dRoots = solveset(df, x)
ddRoots = solveset(ddf, x)
dRoots
```

```python
[_ for _ in dRoots]
value = [f.subs(x, _) for _ in dRoots]
value
```

```python
for i in range(0,len(value)):
    if value[i] > 0:
        print('Given function has relative minima at ',dRoots.args[i])
    elif value[i] < 0:
        print('Given function has relative minima at ',dRoots.args[i])
```
